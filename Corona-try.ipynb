```python
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np
df1=pd.read_csv(r'C:\Users\Adarsh\opofcorona.csv')
print(df1.shape)

geolocator = Nominatim(user_agent="adsreg")
lat = []
lon = []
add =[]

for num,s in df1[df1['num']>8830].iterrows():
 #i=s['Location']    
 location = geolocator.geocode(s['Location'], timeout=10)
 #print("Latitude and Longitude Values:")
 located=str(location)
 count=0
 if(not(str(location).isdigit())):
    count+=1
    if(hasattr(location,'latitude')):
    #try: 
    
     #print(location,location.address,location.latitude, location.longitude)
     lat.append(location.latitude)
     lon.append(location.longitude)
     addr=location.address.split(',')
     
     #print(addr)
     add.append(addr[-1])
    else:
     lat.append(np.nan)
     lon.append(np.nan)
     add.append(np.nan)
 else:
     lat.append(np.nan)
     lon.append(np.nan)
     add.append(np.nan)
```
